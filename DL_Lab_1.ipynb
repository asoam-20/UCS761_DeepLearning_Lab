```python
def predict(x1,x2,w1,w2,b):
  z = (x1*w1)+(x2*w2)+b

  return 1 if z>=0 else 0

def perceptron(data,learning_rate = 0.1,epochs = 10):
  w1 = 0.5
  w2 = 0.1
  b = -0.2

  print(f"Undergoing perceptron training:")

  for epoch in range(epochs):
      total_error = 0

      for x1,x2,y in data:
        y1 = predict(x1,x2,w1,w2,b)
        error = y - y1

        if error != 0:
          w1 += learning_rate*error*x1
          w2 += learning_rate*error*x2
          b += learning_rate*error
          total_error += abs(error)

      if total_error == 0:
       print("Converged")
       break

  return w1,w2,b

```
